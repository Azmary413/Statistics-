```python
person = c(10,5,5,10,10)
sports =c("Footbal","Hockey","Cricket","Basketball","Badminton")
```

```python
pie(person, sports, main ="Sports Percentage", col=rainbow(length(person)))
legend("right", sports, cex= .60, fill=rainbow(length(person)))
```
